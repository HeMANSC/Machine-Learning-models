```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

### **Making meaningfull dataset**

```python
df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df.head()
```

```python
df['target'] = breast_cancer.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['target'].value_counts()
```

* Data is **Imbalenced**

**1-->Benign Cancre**

**0-->Malignant Cancre**

### **Seprating the features and target**

```python
x = df.drop(columns='target', axis=1)
y = df['target']
```

### **Spliting the data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

### **Standrize the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scal = scaler.fit_transform(x_train)
x_test_scal = scaler.transform(x_test)
```

### **Bulding the Neural Network**

```python
df.shape
```

```python
# Import the tensorflow and keras
import tensorflow as tf
from tensorflow import keras
```

```python
# Setting up layers for NN
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), # It use to to convert multi-dimensional input data into a one-dimensional vector.
                          keras.layers.Dense(20,activation = 'relu'),
                          keras.layers.Dense(20,activation = 'relu'),# Dense layer is a fully connected neural network layer where each neuron is connected to all neurons in the previous layer
                          keras.layers.Dense(2,activation = 'sigmoid')

                          ])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_scal, y_train, validation_split=0.1,epochs=10)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred = model.predict(x_train_scal)
```

```python
y_pred[3]
```

```python
train_pred = model.predict(x_train_scal)
train_pred = [np.argmax(i) for i in  train_pred]
print(train_pred)
```

```python
print('train_acc :- ',accuracy_score(y_train, train_pred))
```

```python
test_pred = model.predict(x_test_scal)
test_pred = [np.argmax(i) for i in  test_pred]
print(test_pred)
```

```python
print('test_acc :- ',accuracy_score(y_test, test_pred))
```
